```python
#  STUDENT NAME
#  CSC 201 - SPRING 2019
#  DICTIONARIES AND PANDAS DATAFRAMES
#  DATE 4/4/19
   
#  PROGRAM TITLE: YouTube Data PROGRAM
   
#  PROGRAM DESCRIPTION: A Python program that opens and reads the data from the CSV file into a 
#                      
                        
#  GENERAL SOLUTION: A Python program that opens and reads the data from the CSV file into a 
#                       Pandas DataFrame and uses this DataFrame to output information from it
#                       based on the user's choice. The choices in this program include:
#                             output information available in DataFrame on a channel's rank
#                             output information available in DataFrame on a channel's grade
#                             Quit the program.

#  PSEUDOCODE: 

import pandas as pd

# function name: read_as_dataframe
# arguments: 1 string (represents the path and filename of the CSV file)
# return: 1 pandas dataframe
# description: reads in each row of the CSV file as a pandas dataframe and
#              returns this dataframe with the data from the CSV file 
def read_as_dataframe(filename): 
    df = pd.read_csv(filename)
    #df = pd.read_csv(filename, error_bad_lines=False)

    
    return df


# func name: is_integer
# arguments: 1 value (any type)
# return: 1 boolean (true if integer, false if not integer)
# description: check if value is an integer
def is_integer(s): 
    try:
        int(s)
        return True
    
    except ValueError: 
        return False


# function name: get_rank
# arguments: 1 pandas dataframe
# return: either a string stating unable to find error OR the filtered subset on the rank of a channel
# description: asks the user to input a channels rank and check if this rank is in the passed
#              pandas dataframe
#                     if the rank is not in the dataframe, then return string saying unable to find
#                     otherwise, return the filtered subset on the rank 
def get_rank(df):
    rank = int(input("Enter the channel rank:   "))
    is_rank = df[(df["Rank"] == rank)]
    
    
    if is_integer(rank) == False:
        return "The rank must contain only digits. "


    if is_rank.empty :
        return "Unable to find a channel with that Rank. Try again."
        
    return is_rank
        
    
# function name: get_grade
# arguments: 1 pandas dataframe
# return: either a string stating unable to find error OR the filtered subset on the grade of a channel
# description: asks the user to input a channels grade and check if this grade is in the passed
#              pandas dataframe
#                     if the grade is not in the dataframe, then return string saying unable to find
#                     otherwise, return the filtered subset on the grade    
def get_grade(df):
    grade = input("Enter the channel's grade:   ")
    is_grade = df[(df["Grade"] == grade)]
    
    if is_grade.empty :
        return "Unable to find a channel with that Grade. Try again."
    
    else:
        return is_grade    

    
# function name: main
# arguments: none
# return: none
# description: handles checking the user input for their choice and calling the other functions when needed 
def main():
    file  = "./data/youtubedata2.csv"
    
    df = read_as_dataframe(file)
    
    choice = ""
    
    while choice != "3" :
        print(''' Choose one of the options below:
        1. Search by Rank.
        2. Search by Grade. 
        3. Quit
        ''')
        
        choice = input("Choice: ")
        print("")


        if choice == "1":
            rank_output = get_rank(df)
            print("")
            print(rank_output)

        elif choice == "2":
            grade_output = get_grade(df)
            print("")
            print(grade_output)

        elif choice == "3":
            print("Quitting the program...Program has ended.")

        else:
            print("You did not enter a valid choice. Try Again!")

    return



main()
```
