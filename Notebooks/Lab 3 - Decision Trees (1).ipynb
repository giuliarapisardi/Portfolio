```python
# Luca G, Stephanie, Giulia
# Lab #3 - Decision Trees 2.17.20

import pandas as pd
from sklearn import tree
from treeviz import tree_print
from sklearn.metrics import accuracy_score
```

```python
#import data set
df = pd.read_csv('tennis_numeric.csv')
df.head()
```

```python
# create features vector
features_df = df.drop(['play'],axis=1)
features_df.head()
```

```python
# create target vector
target_df = pd.DataFrame(df['play'])
target_df.head()
```

```python
# set up the decision tree model
dtree = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
# create the decision tree model
dtree.fit(features_df,target_df)
```

```python
# show the actual model
tree_print(dtree,features_df)
```

```python
# evaluate the model for accuracy
predict_array = dtree.predict(features_df)
predicted_labels = pd.DataFrame(predict_array)
predicted_labels.columns = ['play']
```

```python
predicted_labels.head()
```

```python
target_df.head()
```

```python
# check if the predicted labels are equal to the labels in the target vector
predicted_labels.equals(target_df)
```

```python
# check the accuracy score of the model
from sklearn.metrics import accuracy_score
print('Our model accuracy is: {}'.format(accuracy_score(target_df,predicted_labels)))
```

```python
#import data set
df2 = pd.read_csv('Crackers.csv')
df2.drop(['id'], axis = 1, inplace = True)
```

```python
# create features vector
features_df2 = df2.drop(['choice'],axis=1)
features_df2.head()
```

```python
# create target vector
target_df2 = pd.DataFrame(df2['choice'])
target_df2.head()
```

```python
# set up the decision tree model
dtree2 = tree.DecisionTreeClassifier(criterion = 'entropy')
```

```python
# create the decision tree model
dtree2.fit(features_df2,target_df2)
```

```python
# evaluate the model for accuracy
predict_array2 = dtree2.predict(features_df2)
predicted_labels2 = pd.DataFrame(predict_array2)
predicted_labels2.columns = ['choice']
predicted_labels2.head()
```

```python
# show the actual model
tree_print(dtree2,features_df2)
```

```python
target_df2.head()
```

```python
predicted_labels2.equals(target_df2)
```

```python
# check the accuracy score of the model
from sklearn.metrics import accuracy_score
print("Our model accuracy is: {}".format(accuracy_score(target_df2, predicted_labels2)))
```

```python
# supplemental tree, restriced the depth to see how the model would change

dtree3 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=8)
dtree3.fit(features_df2,target_df2)
```

```python
tree_print(dtree3,features_df2)
```

```python
# evaluate the model for accuracy
predict_array3 = dtree3.predict(features_df2)
predicted_labels3 = pd.DataFrame(predict_array3)
predicted_labels3.columns = ['choice']
predicted_labels3.head()
```

```python
# check the accuracy score of the model
from sklearn.metrics import accuracy_score
print("Our model accuracy is: {}".format(accuracy_score(target_df2, predicted_labels3)))
```
