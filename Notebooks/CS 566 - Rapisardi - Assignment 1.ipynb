1. Use python to create 3 different plots of the following functions
    
    $f_{1}(n)$ = $2^{10}(n)$+ $2^{10}$
    
    $f_{2}(n)$ = $n^{3.5}$ - $1000$
    
    $f_{3}(n)$ = $100n^{2.1}$ + $50$


(a) Create 3 plots and limit the horizontal x-axis to n = 5, 15, 50. On each of the 3 plots you need to show the above 3 functions. On the first plot the x-axis is limited to 5, on second one x-axis is limited to 15 and on the 3rd one x-axis is limited to 50.

(b) Visualize the 3 functions in 3 colors ( $f_{1}$ in red, $f_{2}$ in blue, $f_{3}$ in green).


```python
# import necessary packages
import math
import numpy as np
import matplotlib.pyplot as plt
```

graphing the functions with the limit n = 5 on the x-axis

```python
msize = 5

t = np.arange(0, msize, 0.1)

plt.plot(t, 2**10*t + 2**10, 'red', t, t**3.5 - 1000, 'blue', t, 100*t**2.1, 'green')

plt.xlim(0, msize)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```

graphing the functions with the limit n = 15 on the x-axis

```python
msize = 15

t = np.arange(0, msize, 0.1)

plt.plot(t, 2**10*t + 2**10, 'red', t, t**3.5 - 1000, 'blue', t, 100*t**2.1, 'green')

plt.xlim(0, msize)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```

graphing the functions with the limit n = 50 on the x-axis

```python
msize = 50

t = np.arange(0, msize, 0.1)

plt.plot(t, 2**10*t + 2**10, 'red', t, t**3.5 - 1000, 'blue', t, 100*t**2.1, 'green')

plt.xlim(0, msize)
plt.rcParams["figure.figsize"] = (7,7)
plt.show()
```

(c) Describe your visualization and what you see in these 3 plots.

The first graph displays the x-axis with a limit n = 5 where immediately $f_{1}$ shows exponential growth compared to the other two graphs. Once the n value increased to 15 the function $f_{3}$ grows much faster than both $f_{1}$ and $f_{2}$. Finally, the limit of the x-axis is increased to n = 50 where we can see that the exponential growth of $f_{2}$ is not prominent in comparison to the two functions until about n = 27. From there on $f_{2}$ grows exponentially and at a much faster rate than the other two functions.
